```python
import pandas as pd
file = 'data/2021/passing.csv'
```

```python
def formatNames(player):
    if '*' in player:
        player = player.split('*')[0]
    else:
        player = player.split('\\')[0]
    return player

data = pd.read_csv(file)
data['Player'] = list(map(formatNames, list(data['Player'])))
data = data.drop(data.columns[0], axis=1)
data
```

```python
data.to_csv(file)
```
