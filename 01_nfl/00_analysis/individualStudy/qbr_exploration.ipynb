```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
file = 'data/2021/passing.csv'
```

```python
data = pd.read_csv(file)
data = data.drop(data.columns[0], axis=1)
data
```

#### QBR Average vs Playoff QBs QBR Average for Each Round
Do better QBs advance further in the playoffs?

```python
qbrComparison = data[['Player','Cmp','Att','Yds','TD','Int','Rate','QBR','Sk']]
qbrComparison = qbrComparison.loc[qbrComparison['Att'] > 100]
qbrComparison['Round'] = 0
playoffQBList = ['Derek Carr','Joe Burrow',
                 'Mac Jones','Josh Allen',
                 'Jalen Hurts','Tom Brady',
                 'Jimmy Garoppolo','Dak Prescott',
                 'Ben Roethlisberger','Patrick Mahomes',
                 'Kyler Murray','Matthew Stafford',
                 'Ryan Tannehill','Aaron Rodgers'
                ]
playoffQBs = qbrComparison.loc[qbrComparison['Player'].isin(playoffQBList)].set_index('Player').reset_index()
playoffQBs['Round'] = [2,3,3,1,3,1,2,2,3,1,1,1,2,1]

# Adding Rounds back to the original qbrComparison and exporting for charting.
qbrComparison = pd.merge(qbrComparison,playoffQBs[['Player','Round']],on='Player', how='left')
qbrComparison['Round'] = qbrComparison['Round_y']
qbrComparison = qbrComparison.drop(columns=['Round_x', 'Round_y'])
qbrComparison['Round'] = qbrComparison['Round'].fillna(0).astype(int)
qbrComparison.to_csv('data/forTooltipCharting/0001.csv')
```

##### Wild Card, DIvisional, Conference Winners

```python
wildCardWinners = playoffQBs.loc[playoffQBs['Round'] > 0]
divisionalWinners = wildCardWinners.loc[wildCardWinners['Round'] > 1]
```

##### Average QBR, Passer Rating

```python
image = playoffQBs.sort_values('QBR', ascending=False).set_index('Player').reset_index()
image = image[['Player','Yds','TD','Int','Rate','QBR']]
image
```

```python
averageQBR = round(qbrComparison['QBR'].mean(), 1)
playoffQBaverageQBR = round(playoffQBs['QBR'].mean(), 1)
wildcardWinnersAverageQBR = round(wildCardWinners['QBR'].mean(), 1)
divisionalWinnersAverageQBR = round(divisionalWinners['QBR'].mean(), 1)

averageRate = round(qbrComparison['Rate'].mean(), 1)
playoffQBaverageRate = round(playoffQBs['Rate'].mean(), 1)
wildcardWinnersAverageRate = round(wildCardWinners['Rate'].mean(), 1)
divisionalWinnersAverageRate = round(divisionalWinners['Rate'].mean(), 1)

results = pd.DataFrame({
    '': ['League','Playoff','WildCardWinners','DivisionalWinners'],
    'QBR': [averageQBR, playoffQBaverageQBR, wildcardWinnersAverageQBR, divisionalWinnersAverageQBR],
    'Rate': [averageRate, playoffQBaverageRate, wildcardWinnersAverageRate, divisionalWinnersAverageRate]
}).set_index('')
results
```
