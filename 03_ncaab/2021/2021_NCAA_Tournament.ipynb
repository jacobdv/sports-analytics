```python
# Formatting tables for html.
ncaa = ncaa.drop(["Rank"],axis=1)
ncaa_adv = ncaa_adv.drop(["Rank"],axis=1)
ncaa_opp = ncaa_opp.drop(["Rank"],axis=1)
ncaa_opp_adv = ncaa_opp_adv.drop(["Rank"],axis=1)
```

```python
ncaa["W/L%"] = round(ncaa["W/L%"] * 100,1)
ncaa_adv["W/L%"] = round(ncaa_adv["W/L%"] * 100,1)
ncaa_adv["TS%"] = round(ncaa_adv["TS%"] * 100,1)
ncaa_adv["eFG%"] = round(ncaa_adv["eFG%"] *100,1)
ncaa_opp["W/L%"] = round(ncaa_opp["W/L%"] * 100,1)
ncaa_opp_adv["W/L%"] = round(ncaa_opp_adv["W/L%"] * 100,1)
ncaa_opp_adv["TS%"] = round(ncaa_opp_adv["TS%"] * 100,1)
ncaa_opp_adv["eFG%"] = round(ncaa_opp_adv["eFG%"] * 100,1)
```

```python
def remove_NCAA(school):
    school = school.replace(" NCAA","")
    return school
```

```python
schools_list = []
for school in ncaa["School"]:
    school = remove_NCAA(school)
    schools_list.append(school)
ncaa["School"] = schools_list
```

```python
# Key Stats
cs = pd.DataFrame()

# Pulls Additional Stats
cs["School"] = ncaa["School"]
cs["Games"] = ncaa["Games"]
cs["ORS%"] = round(100 * (ncaa["ORB"] / (ncaa["ORB"] + (ncaa_opp["TRB"] - ncaa_opp["ORB"]))),2)
cs["DRS%"] = round(100 * (ncaa["TRB"] - ncaa["ORB"]) / ((ncaa["TRB"] - ncaa["ORB"]) + ncaa_opp["ORB"]),2)
cs["TRS%"] = cs["ORS%"] + cs["DRS%"]
cs["A%"] = round((ncaa["AST"] / ncaa["FG"]),2)
cs["Pace"] = ncaa_adv["Pace"]
cs["FGA"] = ncaa["FGA"]
cs["FG%"] = ncaa["FG%"]
cs["FT/FGA"] = ncaa_adv["FT/FGA"]
cs["FT%"] = ncaa["FT%"]
cs["FT"] = ncaa["FT"]
cs["FTA"] = ncaa["FTA"]
cs["3PAr"] = ncaa_adv["3P Attempt Rate"]
cs["2P%"] =  ((ncaa["FG"] - ncaa["3P"]) / (ncaa["FGA"] - ncaa["3PA"]))
cs["3P%"] = ncaa["3P%"]
cs["AST"] = ncaa["AST"]

# Filter Stats
cs["W%"] = round(100 * (ncaa["Wins"]/ncaa["Games"]),1)
cs["SOS"] = ncaa["SOS"]
```

```python
zwaz = pd.DataFrame()
zwaz["School"] = cs["School"]
zwaz["W%"] = cs["W%"]
zwaz["FF"] = 1000 *((ncaa_adv["eFG%"] / 100) * 0.4) * ((ncaa_adv["TOV%"] / 100) * 0.25) * (cs["TRS%"] / 100) * (cs["FT/FGA"])

### Points created from turnovers. ###
# Turnover Differential %
zwaz["TDiff"] = (ncaa_adv["BLK%"] + ncaa_adv["STL%"]) - ncaa_adv["TOV%"]
# Turnover Diff Possessions
zwaz["TDiff"] = ((zwaz["TDiff"] / 100) * cs["Pace"])
# Turnover Diff Points
zwaz["TDiff"] = zwaz["TDiff"] * (ncaa["Team Points"] / (cs["Pace"] * ncaa["Games"]))

### Points created from free throws. ###
zwaz["FT"] = ncaa["FT"] / ncaa["Games"]

### Two point threat. ###
zwaz["T2"] = 2 * (((((ncaa["FGA"] - ncaa["3PA"]) / cs["Pace"]) * st.mean(cs["Pace"])) * cs["2P%"]) / ncaa["Games"])

### Three point threat. ###
zwaz["T3"] = 3 * ((((ncaa["3PA"] / cs["Pace"]) * st.mean(cs["Pace"])) * ncaa["3P%"]) / ncaa["Games"])

### Examines fouls. ###
# Fouls per possession.
zwaz["F"] = ncaa["PF"] / (cs["Pace"] * ncaa["Games"])
# Fouls per national average game.
zwaz["F"] = zwaz["F"] * st.mean(cs["Pace"])
# Opponent free throws per national average game.
zwaz["F"] = (ncaa_opp["FTA"] / ncaa["PF"]) * zwaz["F"]
# Opponent foul points per game.
zwaz["F"] = (zwaz["F"] * st.mean(ncaa["FT%"]))

# Sums points from fts, twos, threes, and the turnover differential.
# Subtracts heavily averaged opponent free throw points.
zwaz["Z-Alg"] = (zwaz["TDiff"] + zwaz["FT"] + zwaz["T3"] + zwaz["T2"]) - zwaz["F"]

zwaz["SRS"] = ncaa["SRS"]
zwaz
```

```python
def team_score (team,measure):
    team_name = team
    stat = str(zwaz[measure].loc[zwaz["School"] == team])
    stat = stat.split()
    stat = stat[1]
    team_dict = {}
    team_dict = {
        "School":team,
        measure:stat
    }
    return team_dict
```

```python
teams = ['Gonzaga','Norfolk State','Oklahoma','Missouri',
        'Creighton','UC-Santa Barbara','Virginia','Ohio',
        'Southern California','Wichita State','Kansas','Eastern Washington',
        'Oregon','Virginia Commonwealth','Iowa','Grand Canyon',
        
        'Michigan','Texas Southern','Louisiana State','St. Bonaventure',
        'Colorado','Georgetown','Florida State','North Carolina-Greensboro',
        'Brigham Young','Michigan State','Texas','Abilene Christian',
        'Connecticut','Maryland','Alabama','Iona',
        
        'Baylor','Hartford','North Carolina','Wisconsin',
        'Villanova','Winthrop','Purdue','North Texas',
        'Texas Tech','Utah State','Arkansas','Colgate',
        'Florida','Virginia Tech','Ohio State','Oral Roberts',
        
        'Illinois','Drexel','Loyola (IL)','Georgia Tech',
        'Tennessee','Oregon State','Oklahoma State','Liberty',
        'San Diego State','Syracuse','West Virginia','Morehead State',
        'Clemson','Rutgers','Houston','Cleveland State']
```

```python
i = 0
measure = "Z-Alg"
while len(teams) > 1:
    i-=1
    # Gets first team in matchup.
    school_one = team_score(teams[i],measure)
    i+=1
    # Gets second team in matchup.
    school_two = team_score(teams[i],measure)
    i+=1
    if school_one[measure] >= school_two[measure]:
        teams.remove(teams[(i-1)])
    elif school_one[measure] < school_two[measure]:
        teams.remove(teams[(i-2)])
    if len(teams) == 32 or len(teams) == 16 or len(teams) == 8 or len(teams) == 4 or len(teams) == 2 or len(teams) == 1:
        i=0
        print(f"\nRound of {2* (len(teams))} Winners:")
        print(teams) 
```
