```python
import pandas as pd
from config import key
import requests
```

```python
# Team Season Stats URL
year = 2020 # Default as the free API includes only current year.
tss_url = f"https://api.sportsdata.io/v3/cbb/scores/json/TeamSeasonStats/{year}?key={key}"
```

```python
# This is the actual limited API call part. (1000 per month for the free trial.)
full_season_stats = requests.get(tss_url).json()
```

```python
# Lists to Make
team_names = []
wins = []
losses = []
conf_wins = []
conf_losses = []
possessions = []
efg = []
two_pm = []
two_pa = []
three_pm = []
three_pa = []
ftm = []
fta = []
orb = []
drb = []
ast = []
stl = []
blk = []
tov = []
pf = []
ts = []
```

```python
#Iterant
i = 0
# Pulls all relevant data.
for team in full_season_stats:
    team_names.append(full_season_stats[i]['Name'])
    wins.append(full_season_stats[i]['Wins'])
    losses.append(full_season_stats[i]['Losses'])
    conf_wins.append(full_season_stats[i]['ConferenceWins'])
    conf_losses.append(full_season_stats[i]['ConferenceLosses'])
    possessions.append(full_season_stats[i]['Possessions'])
    efg.append(full_season_stats[i]['EffectiveFieldGoalsPercentage'])
    two_pm.append(full_season_stats[i]['TwoPointersMade'])
    two_pa.append(full_season_stats[i]['TwoPointersAttempted'])
    three_pm.append(full_season_stats[i]['ThreePointersMade'])
    three_pa.append(full_season_stats[i]['ThreePointersAttempted'])
    ftm.append(full_season_stats[i]['FreeThrowsMade'])
    fta.append(full_season_stats[i]['FreeThrowsAttempted'])
    orb.append(full_season_stats[i]['OffensiveRebounds'])
    drb.append(full_season_stats[i]['DefensiveRebounds'])
    ast.append(full_season_stats[i]['Assists'])
    stl.append(full_season_stats[i]['Steals'])
    blk.append(full_season_stats[i]['BlockedShots'])
    tov.append(full_season_stats[i]['Turnovers'])
    pf.append(full_season_stats[i]['PersonalFouls'])
    ts.append(full_season_stats[i]['TrueShootingPercentage'])
    
    # Iterant
    i = i + 1
```

```python
instinct_stats = pd.DataFrame()
instinct_stats["Team"] = team_names
instinct_stats["Wins"] = wins
instinct_stats["Losses"] = losses
instinct_stats["ORB"] = orb
instinct_stats["DRB"] = drb
instinct_stats["TS%"] = ts
```

```python
instinct_stats = instinct_stats[instinct_stats["Wins"]/instinct_stats["Losses"] > 1.3]
instinct_stats = instinct_stats[instinct_stats["Wins"]+instinct_stats["Losses"] > 20]
```
